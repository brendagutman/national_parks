```python
import pandas as pd
```

```python
parks_data = pd.read_csv("../data/parks_info.csv")
visitor_data = pd.read_csv("../data/recreation_visitors.csv")
parks_data.head(50)
```

https://stackoverflow.com/questions/12680754/split-explode-pandas-dataframe-string-entry-to-separate-rows/57122617#57122617

```python
activities = parks_data[['fullName','activities']]
activities['activities'] = activities['activities'].str.replace('[','')
activities['activities'] = activities['activities'].str.replace(']','')
activities['activities'] = activities['activities'].str.replace('\'','')
activities.assign(activities=activities['activities'].str.split(',')).explode('activities')
```

```python
activities.to_csv('../data/activities.csv')
```
